```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian.csv')
df.head()
```

```python
# drop month day and year
df.drop(['day','month','year'],axis=1,inplace=True)
df.head()
```

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df.head(2)
```

```python
df['Classes'].value_counts()
```

```python
X=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.corr()
```

```python
## Check for multicollinearity

plt.figure(figsize=(12,13))
corr = X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
import numpy as np

def correlation(dataset, threshold):
    
    corr_matrix = dataset.corr(numeric_only=True)
    
    col_corr = set()
    
    for i in range(len(corr_matrix.columns)):
        
        for j in range(i):
            
            if abs(corr_matrix.iloc[i, j]) > threshold:
                
                colname = corr_matrix.columns[i]
                
                col_corr.add(colname)
                
    return col_corr
```

```python
# Threshold -- Domain Expertise
corr_features = correlation(X_train,0.85)
corr_features
```

```python
## Drop features when correlation is more than 0.85
X_train.drop(corr_features,axis=1,inplace=True,errors='ignore')
X_test.drop(corr_features,axis=1,inplace=True,errors='ignore')
X_train.shape,X_test.shape
```

## Feature Scaling or Standardization

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
X_train_scaled = Scaler.fit_transform(X_train)
X_test_scaled = Scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
## Box Plot to understand of Standard Scaler

plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title("X_train Before Scaling")
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title("X_train After Scaling")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , r2_score , mean_squared_error
lr=LinearRegression()
lr.fit(X_train_scaled,y_train)
y_pred = lr.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("MAE",mae)
print("r-2score",score)
plt.figure(figsize=(5,3))
plt.scatter(y_test,y_pred)
plt.show()
```
